# #split the data into features and target

```python
X = dataset.iloc[:,:-1]
X
```

```python
y = dataset.iloc[:,[-1]]
y
```

# #data preprocessing

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
blum = LabelEncoder()
Location = blum.fit_transform(X['Location'])
```

```python
Location
```

```python
X['Location'] = Location
X
```

```python
X.describe()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
```

```python
numeric_columns = ['HouseAge', 'Bedroom', 'FullBath', 'LotArea', 'Location']
transformer = MinMaxScaler()
```

```python
preprocess = ColumnTransformer(
transformers=[
    ('numerical', transformer, numeric_columns)
])
```

```python
# model
base_model = RandomForestRegressor( n_estimators=200, random_state=0)   
```

```python
model = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', base_model)
])
```

# split the dataset into train and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train
```

# #Training the Model

```python
# from sklearn.ensemble import RandomForestRegressor
```

```python
# model = RandomForestRegressor( n_estimators=200, random_state=0)   
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

# #Model Evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2 = r2_score(y_test, prediction)
mae = mean_absolute_error(y_test, prediction)
mse = mean_squared_error(y_test, prediction)

print('r2 score: ', r2)
print('mean absolute error: ', mae)
print('mean_square error: ', mse)
```
